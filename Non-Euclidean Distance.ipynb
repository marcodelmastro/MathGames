```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
def y(x,n=2,c=1):
    return np.power(c**n-x**n,1/n)
```

```python
fig = plt.figure(figsize=(10,10))
fig,ax = plt.subplots(dpi=100,figsize=(5,5))

n = 4
color = [ 'green', 'red', 'blue', 'yellow', 'purple' ]

for e in range(-20,20):
    n = 1.2**e
    #print(e,n)
    plt.plot( x, y(x,n),color=color[e%len(color)])
    plt.plot(-x, y(x,n),color=color[e%len(color)])
    plt.plot(-x,-y(x,n),color=color[e%len(color)])
    plt.plot( x,-y(x,n),color=color[e%len(color)])
```

```python
from scipy.integrate import quad

#for e in range(-20,20):
#    n = 1.2**e
for n in range(1,4):
    pi = 4.*quad(y,0,1,args=(n))[0]
    print(n,pi)
```
